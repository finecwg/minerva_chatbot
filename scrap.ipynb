```python
import requests
from bs4 import BeautifulSoup
import pdfkit
import os

# path to wkhtmltopdf executable
path_wkhtmltopdf = '/usr/bin/wkhtmltopdf'
config = pdfkit.configuration(wkhtmltopdf=path_wkhtmltopdf)

def fetch_sitemap(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
        return [loc.text for loc in soup.find_all('loc')]
    except requests.exceptions.RequestException as e:
        print(f"Error fetching sitemap: {e}")
        return []

def save_webpage_as_pdf(url, output_dir):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
        html_file = os.path.join(output_dir, 'temp.html')
        pdf_file = os.path.join(output_dir, f"{url.split('//')[-1].replace('/', '_')}.pdf")
        
        with open(html_file, 'w', encoding='utf-8') as file:
            file.write(soup.prettify())
        
        pdfkit.from_file(html_file, pdf_file)
        print(f"Saved PDF: {pdf_file}")
        #os.remove(html_file)
    except Exception as e:
        print(f"Failed to process {url}: {e}")

def main():
    sitemap_url = 'https://my.minerva.edu/sitemap-cms.xml'
    output_dir = 'pdf_output'

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    pages = fetch_sitemap(sitemap_url)
    for page in pages:
        print(f"Processing: {page}")
        save_webpage_as_pdf(page, output_dir)

```

```python
if __name__=='__main__':
  main()
```
